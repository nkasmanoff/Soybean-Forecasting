```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import os
```

```python
#to start like in 2, let's just manage one 
#dataset at a time
soybean_yield = pd.read_csv('../dat/clean_yield.csv')
fertilizers = pd.read_csv('../dat/fertilizers_cleaned.csv')
insecticides = pd.read_csv('../dat/insecticides_cleaned.csv')
herbicides = pd.read_csv('../dat/herbicides_cleaned.csv')
fungicides = pd.read_csv('../dat/fungicides_cleaned.csv')

climate_data  = pd.read_csv('../../data/climate_df_withtemp.csv')
climate_data = climate_data[climate_data.columns[2:]]
```

```python
panel_stuff = [fertilizers,insecticides,herbicides,fungicides,climate_data]
```

So what do we want to do? See what states and/or years they share. 

```python
panel_df = soybean_yield.copy(deep=True)

for df in panel_stuff:
    panel_df = panel_df.merge(df,on=['YEAR','LOCATION'],how='left')
```

These are states we have yield data for but no fertilizer data for. 

```python
panel_df.to_csv('../../panel_df.csv',index=False)
```
