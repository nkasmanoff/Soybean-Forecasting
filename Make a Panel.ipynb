```python
import pandas as pd
import numpy as np
import os,sys

import matplotlib.pyplot as plt

%matplotlib inline
```

```python


def generalized_cleaning(df):
    
    df_list = []
    for item, g in df.groupby('Data Item'):
        g[item + ' Value'] = g['Value']
        del g['Value']
        g = g[['Year','Week Ending','State', item + ' Value']]
        df_list.append(g)

    for i in range(len(df_list)):
        if i == 0:
            new_df = df_list[0]

        if 0 < i < len(df_list):
            new_df = new_df.merge(df_list[i], on = ['Year','Week Ending','State'])
            
    return new_df
```

```python
moisture = pd.read_csv('rahul/data/soybean_moisture.csv')
applications = pd.read_csv('rahul/data/soybean_applications.csv')
condition = pd.read_csv('rahul/data/soybean_condition.csv')
pests = pd.read_csv('rahul/data/soybean_pestmgmt.csv')
soy_yield = pd.read_csv('rahul/data/soybean_yield.csv')
```

```python
soybean_yield_clean = generalized_cleaning(soy_yield)
applications_clean = generalized_cleaning(applications)
condition_clean = generalized_cleaning(condition)
moisture_clean = generalized_cleaning(moisture)
pests_clean = generalized_cleaning(pests)
```
