# Steps

    Read in csv
  
    Extract Data Item and value
     
    Merge together
       
    Join over similar years
    
    Repeat for new csv
    
    Join all

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python


def generalized_cleaning(df):
    
    df_list = []
    for item, g in df.groupby('Data Item'):
        g[item + ' Value'] = g['Value']
        del g['Value']
        g = g[['Year','Week Ending','State', item + ' Value']]
        df_list.append(g)

    for i in range(len(df_list)):
        if i == 0:
            new_df = df_list[0]

        if 0 < i < len(df_list):
            new_df = new_df.merge(df_list[i], on = ['Year','Week Ending','State'])
            
    return new_df
```

```python
moisture = pd.read_csv('../rahul/data/soybean_moisture.csv')

applications = pd.read_csv('../rahul/data/soybean_applications.csv')

condition = pd.read_csv('../rahul/data/soybean_condition.csv')

pests = pd.read_csv('../rahul/data/soybean_pestmgmt.csv')
```

```python
generalized_cleaning(pests)
```

```python
moisture_df = []
for item, g in moisture.groupby('Data Item'):
    g[item + ' Value'] = g['Value']
    del g['Value']
    g = g[['Year','Week Ending','State', item + ' Value']]
    moisture_df.append(g)
    
for i in range(len(moisture_df)):
    if i == 0:
        new_moisture_df = moisture_df[0]

    if 0 < i < len(moisture_df):
        new_moisture_df = new_moisture_df.merge(moisture_df[i], on = ['Year','Week Ending','State'])
```
