How similar/different are the test samples from training ones/

```python
import pandas as pd
import numpy as np
```

```python
yield_df = pd.read_csv('../rahul/finaldatasets/final_df.csv')
```

```python
train_val_df = pd.read_csv('../rahul/finaldatasets/train.csv')
test_df = pd.read_csv('../rahul/finaldatasets/test.csv')
```

```python

target = ['YIELD_IN_BU_/_ACRE']
X_tv = train_val_df.drop(target,axis=1)
y_tv = train_val_df[target].values.reshape(-1)


X_test = test_df.drop(target,axis=1)
y_test = test_df[target].values.reshape(-1)
```

```python
X_test['label'] = 'test'
```

```python
X_tv['label'] = 'train_val'
```

```python
from sklearn.manifold import TSNE
```

```python
X_tsne = pd.concat([X_tv.drop(columns=['label']),X_test.drop('label',axis=1)],axis=0)
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python

X_tsne = normalize(X_tsne)
```

```python
X_tsne.dropna(axis=1,inplace=True)
```

```python

X_embedded = TSNE(n_components=2).fit_transform(X_tsne)
labels = pd.concat([X_tv['label'],X_test['label']])
```

```python
test = X_embedded[labels.values== 'test'] # bad fit on training data

tv = X_embedded[labels== 'train_val'] # good fit on training data
```

```python
X_embedded_df = pd.DataFrame(X_embedded)
X_embedded_df['labels'] = labels.values
```

```python
import matplotlib.pyplot as plt
#plt.figure(figsize=(20,20))
plt.plot(X_embedded_df.loc[X_embedded_df.labels == 'test'][0],
         X_embedded_df.loc[X_embedded_df.labels == 'test'][1],'bo',label = 'test data',alpha=.2,markersize=16)
plt.plot(X_embedded_df.loc[X_embedded_df.labels == 'train_val'][0],
        X_embedded_df.loc[X_embedded_df.labels == 'train_val'][1],'ro',label = 'train data',alpha=.2,markersize=16)

plt.legend()
```
