```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Top 10 Commodities exported by USA

```python
df1 = pd.read_csv('USA_comm_data_12-3-2019.csv')
```

```python
df1.plot.bar(y='Value',x='Item')
plt.xlabel('Export commodity')
plt.ylabel('Quantity in tonnes')
```

# Top 10 Commodities exported by Brazil

```python
df2 = pd.read_csv('Brazil_comm_data_12-3-2019.csv')
df2.plot.bar(y='Value',x='Item')
plt.xlabel('Export commodity')
plt.ylabel('Quantity in tonnes')
plt.title('Top 10 Commodities exported by Brazil')
plt.savefig('Top 10 Commodities exported by Brazil')
```

# Top 10 Commodities Produced by USA

```python
df3 = pd.read_csv('US_soybean_prod_data_12-3-2019.csv')
df3.plot.bar(y='Value',x='Item')
plt.xlabel('Commodity')
plt.ylabel('Quantity in tonnes')
plt.title('Top 10 Commodities produced by USA')
plt.savefig('Top 10 Commodities produced by USA')
```

# Soybean production US vs Brazil


```python
df4 = pd.read_csv('US_soybean_data_12-3-2019.csv')
y1 = df4['Value']
x1 = df4['Year']
```

```python
df5 = pd.read_csv('Brazil_soybean_data_12-3-2019.csv')
y2 = df5['Value']
x2 = df5['Year']
```

```python
df5
```

```python
sns.set(font='Arial',style='whitegrid',font_scale=1.5)
```

```python
sns.set(font='Arial',style='whitegrid',font_scale=1.5)
plt.figure(figsize=(8,5))
plt.plot(x1,y1,'bo-',linewidth=2,label = 'United States')
plt.plot(x2,y2,'go-',linewidth=2, label = 'Brazil')
plt.xlabel('Year')
plt.ylabel('Soybean Produced (hg/ha)')
plt.legend(loc = 'lower right')
plt.yscale('log')
```

# Soybeans export USA vs Brazil

```python
df6 = pd.read_csv('USA_soybean_export_data_12-3-2019.csv')
y1 = df6['Value']
x1 = df6['Year']

df7 = pd.read_csv('Brazil_soybean_export_data_12-3-2019.csv')
y2 = df7['Value']
x2 = df7['Year']
```

```python
df7
```

```python
sns.set(font='Arial',style='whitegrid',font_scale=1.5)
plt.figure(figsize=(8,5))
plt.plot(x1,y1,'bo-',label = 'United States',linewidth=2)
plt.plot(x2,y2,'go-', label = 'Brazil',linewidth=2)
plt.xlabel('Year')
plt.ylabel('Amount Exported (Tonnes)',fontsize=20)
plt.yscale('log')
plt.legend(loc = 'lower right')
```
