# Read in climate data from the general url above. 

# Steps: First read from the url the actual measurements, which means skipping the first 4 rows. 

# Next, read in just the first row, which contains the name of the state. 

```python
import numpy as np
import pandas as pd
```

```python
state_num = '1'
datatype = 'pmdi' #precipitation. Other examples include Tmax, Tavg, etc.
url = 'https://www.ncdc.noaa.gov/cag/statewide/time-series/'+state_num+'-'+datatype+'-all-12-1996-2019.csv?base_prd=true&begbaseyear=1901&endbaseyear=2000'
climate_df = pd.read_csv(url,skiprows=3)
state = pd.read_csv(url,nrows=0).columns[0]
climate_measure = pd.read_csv(url,nrows=0).columns[1]
print(state)
climate_df['Date'] = climate_df['Date'].astype(str)
climate_df['YEAR'] = climate_df['Date'].apply(lambda z: z[0:4])
climate_df['MONTH'] = climate_df['Date'].apply(lambda z: z[4:])
climate_df['LOCATION'] = [state.upper() for i in range(len(climate_df))]
#climate_df['Date'] = climate_df['YEAR']+'-'+climate_df['MONTH']
climate_df.rename(columns={'Value':'PMDI'},inplace=True)

#now just re arrange

climate_df = climate_df[['LOCATION','YEAR','MONTH','PMDI','Anomaly']]
```

```python
for i in range(2,50):
    state_num = str(i) #example number
    datatype = 'pmdi' #precipitation. Other examples include Tmax, Tavg, etc.
    url = 'https://www.ncdc.noaa.gov/cag/statewide/time-series/'+ state_num +'-'+ datatype +'-all-12-1996-2019.csv?base_prd=true&begbaseyear=1901&endbaseyear=2000'
    climate_df1 = pd.read_csv(url,skiprows=3)
    state = pd.read_csv(url,nrows=0).columns[0]
    #climate_measure = pd.read_csv(url,nrows=0).columns[1]    
    climate_df1['Date'] = climate_df1['Date'].astype(str)
    climate_df1['YEAR'] = climate_df1['Date'].apply(lambda z: z[0:4])
    climate_df1['MONTH'] = climate_df1['Date'].apply(lambda z: z[4:])
    climate_df1['LOCATION'] = [state.upper() for i in range(len(climate_df1))]
    climate_df1.rename(columns={'Value':'PMDI'},inplace=True)
    climate_df1 = climate_df1[['LOCATION','YEAR','MONTH','PMDI','Anomaly']]
    
    climate_df = pd.concat([climate_df,climate_df1],axis=0)

    print(state)
```
